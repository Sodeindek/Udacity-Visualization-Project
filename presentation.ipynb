```python
#importing the dataset and looking through its structure

df = pd.read_csv(r"C:\Users\HP\Desktop\Python\Ford GoBike\ford Gobike.csv")
df.head()
```

## Visualisation 1
## User Gender Distribution

This is a univariate exploration, it shows the distribution of start hours of riders. Weekends are less busier than the rest of the week days

```python
def mycounting(df,xVar, hue=None, color=0, palette=None,order=None, hue_order=None) :
    ''' This me practicing out looping to avoid repetitive codes'''
    #set plot dimension
    plt.figure(figsize=[14,7])
    #plotting
    sb.countplot(data=df,x=xVar, hue=hue,color=sb.color_palette()[color],palette=palette,order= order,edgecolor='black',hue_order=hue_order,linewidth=2);
    # title and labels
    xVar= xVar.replace('_',' ')
    if hue: 
        hue=hue.replace('_',' ')
    #Adding title
    plt.title(f'''Distribution of {xVar} {'by' if hue else ''} {hue if hue else''}'''.title(), fontsize =14, weight ='bold')
    plt.xlabel(xVar.title(), fontsize=10, weight='bold')
    plt.ylabel('Frequency'.title(), fontsize=10, weight='bold')
    
    #plot start hour
mycounting(df,'start_hour')
```

## Visualisation 2
## Reltionship between user type and trip duration

The following visualisation is a bivariate exploration, it shows the distribution of the various user type and its relationship with trip duration in minutes.

```python
sb.catplot(x='user_type', y='duration_min', data=df, kind='bar', ci=None, edgecolor='black')
plt.title('Average Duration(min) by User type',fontsize=15)
plt.ylabel('Duration(min)',fontsize=13)
plt.xlabel('User Type',fontsize=13)
```

## Visualisation 3
## Correllation between all parameters

This third visualisation is a multivariate, it displays the correllation that exists between specific numerical parameters in the dataset.

```python
data_stat = ['age_diff', 'duration_min',"duration_sec", "member_birth_year"]
plt.figure(figsize = (12,5), dpi = 200)

sb.heatmap(df[data_stat].corr(), annot = True, fmt = '.2f', cmap = 'vlag_r', center = 0)
plt.title(" What is the  Correlation between each variable in the dataset?", y=1.09, fontsize=15, weight='bold');
```
